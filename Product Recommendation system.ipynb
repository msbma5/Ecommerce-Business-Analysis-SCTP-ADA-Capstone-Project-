A product recommendation system is a system whose objective is to predict and compile a list of items that a customer is likely to buy. There are 2 ways to develop a list of recommendation.

### 1. Collaborative filtering
The collaborative filtering method is based on users' previous behaviors, such as the pages they saw, the products they bought, or the ratings they gave to different items. The collaborative filtering method uses this data to find similarities between users or articles, and recommends the most similar articles or content to users.

The basic assumption behind the collaborative filtering method is that those who have seen or bought similar content or products in the past are likely to see or buy similar types of content or products in the future.

Thus, based on this assumption, if one person bought items A, B, and C and another person bought items A, B, and D in the past, it is likely that the first person will buy item D and the other person will buy item C, since they share many similarities with each other

### 2. Content-based filtering
On the other hand, content-based filtering produces a list of recommendations based on the characteristics of an article or user. It usually examines the keywords that describe the characteristics of an article. The basic assumption of the content-based filtering method is that users are likely to see or buy items similar to those they have bought or seen in the past.

A collaborative filtering algorithm is used to recommend products based on user behavior history and similarities between them. The first step in implementing a collaborative filtering algorithm for a product recommendation system is to build a user-to-item matrix.

A user-to-item matrix comprises individual users in the rows and individual elements in the columns.

## Import Data

```python
import pandas as pd
```

```python
df = pd.read_csv('purchase.csv')
```

```python
df.head()
```

```python
df.info()
```

We need to have data where each record contains information about which item each customer has purchased.
We are going to transform the data into a user-to-item matrix, where each row represents one customer and the columns correspond to different products

```python
# Creating a customer to item matrix
customer_item_matrix = df.pivot_table(
    index='customer_id', 
    columns='product_name', 
    values='quantity',
    aggfunc='sum'
)
```

```python
#remove limit to the number of column displayed
pd.set_option('display.max_columns', None)
customer_item_matrix.head()
```

The table above shows the quantity of product per customer. We will now change it to a boolean value, where 1 indicates purchased and 0 is never bought before.

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
```

```python
customer_item_matrix
```

## Collaborative filtering
We will explore two approaches to building our user-based and product-based recommender. In the user-based approach, we calculate the similarities between users based on their item purchase history. In the product-based approach, on the other hand, we calculate the similarities between products based on which products are often purchased along with which other products.

To measure the similarity between users , we will use the cosine_similarity method in the scikit-learn package. 

This cosine_similarity function in the sklearn package calculates the cosine similarities in pairs in the given data.



## User-based collaboration filters and recommendations
To build a user-based collaborative filtering algorithm, we need to calculate the cosine similarities between users. Let's take a look at the following code:

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_user_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix)
)
```

```python
user_user_sim_matrix
```

```python
user_user_sim_matrix.columns = customer_item_matrix.index
```

```python
user_user_sim_matrix
```

```python
user_user_sim_matrix['CustomerID'] = customer_item_matrix.index
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')
```

```python
user_user_sim_matrix
```

```python
# display which customer have similar purchase to CS00005
user_user_sim_matrix.loc['CS00005'].sort_values(ascending=False)
```

```python
# to list out all items bought by A; which is CS0005
items_bought_by_A = set(customer_item_matrix.loc['CS00005'].iloc[
    customer_item_matrix.loc['CS00005'].to_numpy().nonzero()
].index)
```

```python
items_bought_by_A
```

```python
#Select a customer that is similar to CS00005 but not too similar.
#For this case, CS00733 appears 5th in the list so its suitable
items_bought_by_B = set(customer_item_matrix.loc['CS00733'].iloc[
    customer_item_matrix.loc['CS00733'].to_numpy().nonzero()
].index)
```

```python
items_bought_by_B
```

```python
items_to_recommend_to_B = items_bought_by_A - items_bought_by_B
```

```python
items_to_recommend_to_B
```

```python
#to display in table format
df.loc[
    df['product_name'].isin(items_to_recommend_to_B), 
    ['product_name', 'description']
].drop_duplicates().set_index('product_name')
```

# Item based 

```python
item_item_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix.T)
)
```

```python
item_item_sim_matrix
```

```python
item_item_sim_matrix.columns = customer_item_matrix.T.index

item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
```

```python
item_item_sim_matrix
```

```python
top_10_similar_items = list(
   item_item_sim_matrix\
        .loc['Gaming Console']\
        .sort_values(ascending=False)\
        .iloc[:10]\
    .index
)
```

```python
top_10_similar_items
```

```python
df.loc[
    df['product_name'].isin(top_10_similar_items), 
    ['product_name', 'description']
].drop_duplicates().set_index('product_name').loc[top_10_similar_items]
```
