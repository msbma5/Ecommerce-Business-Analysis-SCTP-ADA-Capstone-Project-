# Importing Data

```python
#importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing first dataset ie. Purchase dataset
df = pd.read_csv("purchase.csv")
```

```python
# Taking a quick look at it
display (df.head())
display (df.tail())
display (df.shape)
```

```python
#importing second dataset ie. Customer dataset
df2 = pd.read_csv('customer.csv')
```

```python
# Taking a quick look of it
display (df2.head())
display (df2.tail())
display (df2.shape)
```

# Merging datasets

```python
dfMerged = pd.merge(df , df2, on='customer_id', how= 'outer')
```

```python
display (dfMerged.head())
display (dfMerged.tail())
```

```python
# Data Understanding
display (dfMerged.size)
display (dfMerged.shape)
display (dfMerged.info())
display (dfMerged.columns.values)
```

### Observations
1. From this, we know that there is 1 customer without any purchase/orders yet. We will have to remove it later.

# Data Preparation (Cleaning and Transformation)

## 1.Dealing with Duplicates, Null Values, and Data Types

```python
# Checking if there is any null value
dfMerged.isnull().sum()
```

```python
# Drop values that are null
dfMerged = dfMerged.dropna()
dfMerged
```

```python
# Searching for any duplicates 
dfMerged.duplicated().sum()
```

```python
# Changing Dtypes
# Convert 'order_date' and 'shipping_date' to datetime format
dfMerged['order_date'] = pd.to_datetime(dfMerged['order_date'], dayfirst=True)
dfMerged['shipping_date'] = pd.to_datetime(dfMerged['shipping_date'], dayfirst=True)
```

```python
dfMerged.info()
```

## 2.Adding New Columns

```python
# To display all column names and not truncate it.
pd.set_option('display.max_columns', None)
```

```python
# Extract year, month, day, and day name from 'order_date'
dfMerged['year'] = dfMerged['order_date'].dt.year
dfMerged['month_name'] = dfMerged['order_date'].dt.month_name()
dfMerged['day_name'] = dfMerged['order_date'].dt.day_name()
dfMerged['day'] = dfMerged['order_date'].dt.day
dfMerged['month'] = dfMerged['order_date'].dt.month

# Display the updated DataFrame
display (dfMerged.head(10))
```

```python
# Calculate the price after discount
dfMerged['price_after_discount'] = dfMerged['price'] * (1 - dfMerged['discount'])
dfMerged.head()
```

## 3. Removing irrelevant data

```python
#checking the distribution of orders
# Count the occurrences of each year
year_distribution = dfMerged['year'].value_counts()

# Sort the years in ascending order
year_distribution = year_distribution.sort_index()
```

```python
# Removing data with order_date in 2023
dfMerged.drop(dfMerged[dfMerged['year'] == 2023].index, inplace = True)

# Update the year distribution
updated_year_distribution = dfMerged['year'].value_counts()
updated_year_distribution = updated_year_distribution.sort_index()
```

```python
display(year_distribution)
display(updated_year_distribution)
```

```python
#Final look at the dataset
display (dfMerged.size)
display (dfMerged.shape)
display (dfMerged.info())
display (dfMerged.columns.values)
```

### 2.Multivariate Analysis

1. Who are our top 10 customers based on total sales made?

```python
# Calculate the sales per customer_id
sum_price = dfMerged.groupby('customer_id')['price_after_discount'].sum()
```

```python
# Get the top 10 customers with the highest sum of price
top_10_sales = sum_price.nlargest(10)
top_10_sales
```

```python
# Create the bar chart
plt.figure(figsize=(15,8))
ax = sns.barplot(x=top_10_sales.index, y=top_10_sales.values,color='green')

# Display the amount on top of each column
for i, v in enumerate(top_10_sales.values):
    formatted_value = "${:,}".format(round(v))
    ax.text(i, v, formatted_value, ha='center', va='bottom')
    
# Chart labels
plt.xlabel('Customer ID')
plt.ylabel('Sales')
plt.title('Top 10 Performing Customers based on Sales')
plt.show()
```

2. Who are our top 10 customers based on quantity purchased?

```python
# Calculate quantity per customer_id
sum_quantity = dfMerged.groupby('customer_id')['quantity'].sum()
```

```python
# Get the top 10 customers with the highest sum of price
top_10_quantity = sum_quantity.nlargest(10)
top_10_quantity
```

```python
# Create the bar chart
plt.figure(figsize=(15,8))
ax = sns.barplot(x=top_10_quantity.index, y=top_10_quantity.values,color='orange')

# Display the amount on top of each column
for i, v in enumerate(top_10_quantity.values):
    ax.text(i, v, str(round(v)), ha='center', va='bottom')


# Chart labels
plt.xlabel('Customer ID')
plt.ylabel('Quantity')
plt.title('Top 10 Performing Customers based on Quantity')

plt.show()
```

### 3. Time Series Analysis

```python
# Define the order of weekdays
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
              'August', 'September', 'October', 'November', 'December']
```

1. Which time of year produce the most sales?

```python
sales_by_month = dfMerged.groupby('month_name')['price_after_discount'].sum()
```

```python
# Reindex the sales data based on the defined monthly order
sales_by_month = sales_by_month.reindex(month_order)

# Plot the sales by month
plt.figure(figsize=(15,8))
ax = sns.lineplot(x=sales_by_month.index, y=sales_by_month.values,color='green')


# Y axis number formatting
plt.ticklabel_format(style='plain', axis='y')
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])


plt.ylabel('Sales')
plt.title('Sales by Month')
plt.show()
```
